### Disadvatage of simple moving average

* It is giving the same importance to all data
* It should give more importance to recent data than older data

# Cummulative moving average

eg.  <br>
 12 => 12/1 <br>
 13 => (12 + 13)/2 <br>
 11 => (12 + 13 + 11)/3 <br>
 15 => (12 + 13 + 11 + 15)/4 <br>
 16 => (12 + 13 + 11 + 15 + 16)/5 <br>
 and so on.

```python
## Expading
plt.figure(figsize = (12, 5))
plt.plot(df['close'].expanding().mean())
plt.plot(df['close'])
plt.legend(["CMA", "Close"])
plt.show()
```

# Exponential moving average

* aim should be focus on which data should be focus on
* like focus on recent data
* Formula = ((curEMA - prevEMA)*multiplier) + prevEMA<br>

* **multiplier is calculated as: 2/(window size + 1)**
* if window is 5 then multiplier will be 2/(5 + 1) = 0.33

```python
ema_close = df['close'].ewm(span = 5).mean()
```

```python
## Expading
plt.figure(figsize = (16, 5))
plt.plot(ema_close)
plt.plot(df['close'])
plt.legend(["EMA", "Close"])
plt.show()
```

# Exponential weighted moving average

* It is same as EMA but only difference is multiplier is replaced by alpha

* alpha is hyperparameters which is used to provide more weightage to previous or current values

* Eg.
<img src = "https://miro.medium.com/max/1400/1*CFbYaetEWc2rMHIo8O0OqQ.png" width = "400" height = "200" alt = "EWMA">


```python
ewma_close = df['close'].ewm(alpha = 0.1).mean()
```

```python
## Expading
plt.figure(figsize = (16, 5))
plt.plot(ema_close, c = 'blue') ## EMA
plt.plot(ewma_close, c = 'red') ## EWMA
plt.plot(df['close'], c = 'green')
plt.legend(["EMA","EWMA" , "Close"])
plt.title("Plotting of EMA, EWMA and Data")
plt.show()
```

```python
df_air = pd.read_csv("airline_passengers.csv")
df_air.head()
```

```python
df_air.plot()
```

```python
## It is a seasonal data
## Importing the dicky fuller test
from statsmodels.tsa.stattools import adfuller
```

```python
"""
H0 : Data is not stationary
H1 : Data is stationary

"""

def adf_test(data):
    result = adfuller(data)
    print(f"ADF statistic value: {result[0]}")
    print(f"p-value : {result[1]}")
    if result[1] <= 0.05:
        print("Null hypothesis is rejected.")
        print("Data is stationary.")
    else:
        print("Null hypothesis is accepted.")
        print("Data is not stationary")
```

```python
adf_test(df_air['Thousands of Passengers'])
```

```python
## Here data is not stationary so
## we will use the Differencing technique
df_air['Differencing'] = df_air['Thousands of Passengers'] - df_air['Thousands of Passengers'].shift(1)
df_air.head()
```

```python
df_air.dropna(inplace = True)
```

```python
df_air.head()
```

```python
adf_test(df_air['Differencing'])
```

```python
## Still data is not stationary
## now find second differencing
df_air["Differencing_1"] = df_air['Differencing'] - df_air['Differencing'].shift(1)
df_air.head()
```

```python
df_air.dropna(inplace=True)
```

```python
adf_test(df_air['Differencing_1'])
```

```python
## In seasonal data we genraly consider 12 Months
df_air['Differencing 12 Month'] = df_air['Thousands of Passengers'] - df_air['Thousands of Passengers'].shift(12)
df_air.head()
```

```python
adf_test(df_air['Differencing 12 Month'].dropna())
## Here in single differencing we found the stationary data
```

```python
## train test split
from datetime import datetime, timedelta
train_dataset_end = datetime(1955, 12, 1)
test_dataset_end = datetime(1960, 12, 1)

train_data = df_air[:train_dataset_end]
test_data = df_air[train_dataset_end + timedelta(days = 1): test_dataset_end]
```

```python
pred_start_date = test_data.index[0]
pred_end_date = test_data.index[-1]
```

```python
test_data
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
train_data
```

```python
model_arima = ARIMA(train_data['Thousands of Passengers'], order = (10, 2, 0))
```

```python
model_arima_fit = model_arima.fit()
```

```python
model_arima_fit.summary()
```

```python
pred_start_date, pred_end_date
```

```python
pred = model_arima_fit.predict(start = pred_start_date, end = pred_end_date)
residual = test_data['Thousands of Passengers'] - pred
```

```python
residual
```

```python
model_arima_fit.resid.plot(kind = 'kde')
```

```python
test_data['Predicted_ARIMA'] = pred
```

```python
test_data[['Predicted_ARIMA', 'Thousands of Passengers']].plot()
```

```python
model_sarimax_fit.summary()
```

```python
model_sarimax_fit.resid.plot(kind = 'kde')
```

```python
pred_sarimax = model_sarimax_fit.predict(start = pred_start_date, end = pred_end_date)
```
