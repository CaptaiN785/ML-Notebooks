```python
import nltk
from nltk.stem import PorterStemmer
```

```python
from nltk.corpus import stopwords
```

```python
## Tokenization -> paragraph into 
# nltk.download('punkt') to use tokeniziing

sentences = nltk.sent_tokenize(paragraph)
```

```python
## Removing special characters
import re
corpus = []
for sentence in sentences:
    result = re.sub('[^a-zA-Z]', ' ', sentence)
    corpus.append(result.lower().split())
```

```python
corpus
```

```python
## Removing the stopwords
stopword = stopwords.words('english')
w_list = []
for word_list in corpus:
    words = []
    for word in word_list:
        if word not in stopword:
            words.append(word)
    w_list.append(words)
```

```python
corpus = w_list
```

```python
corpus
```

```python
## Now applly into stemming
stemmer = PorterStemmer()
stemmed_corpus = []
for words in corpus:
    lst = []
    for word in words:
        lst.append(stemmer.stem(word))
    stemmed_corpus.append(lst)
stemmed_corpus
```

```python
## Lematize the words
lemmatizer = WordNetLemmatizer()
lemmatized_corpus = []
for words in corpus:
    lst = []
    for word in words:
        lst.append(lemmatizer.lemmatize(word))
    lemmatized_corpus.append(lst)
lemmatized_corpus
```

```python
new_corpus = []
for words in stemmed_corpus:
    new_corpus.append(' '.join(words))
```

```python
new_corpus
```
