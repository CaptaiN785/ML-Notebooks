## Model Training

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
```

```python
df = red_df[features]
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot = True, fmt = '1.2f')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['quality', 'fixed acidity', 'pH'], axis = 1)
y = df['quality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
train_pred = lr.predict(X_train)
test_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print( "RMSE of train data: ", mean_squared_error(y_train, train_pred) ** 0.5)
```

```python
print( "RMSE of train data: ", mean_squared_error(y_test, test_pred) ** 0.5)
```

```python
r = np.round(test_pred)
```
